```python
# Machine learning
# We need to train data using historical data
# Split the dta into x and y
# This would be done by extracting the data and storing the data as numpy array
array = sales_data.values
array.shape
X=array[:, 0:3]
X.shape

Y=array[:, 3]
Y.shape
```

```python
# Split the dataset into training the sets
# The training datasewt will be used for training data and will take 70% of the training while 30% will be used for evaluation
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, train_size=0.7, random_state=42)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
# Import scientific algorithms
# will be used to create a model that the will learn from our model

# Decision Tree
from sklearn.tree import DecisionTreeRegressor

# Support Vector Machine
from sklearn.svm import SVR

# Random Forest
from sklearn.ensemble import RandomForestRegressor

# Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor

# Linear Regression
from sklearn.linear_model import LinearRegression

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Cross validation(hyperparameter tuning)
# This validates and picks themodel with the highest accuracy
```

```python
# Model training
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
# g) Predicting Results
# The modle was trained on three features(tv,rado,newspaper)
new_features = [[310,70,150]]
predicted_sales = model.predict(new_features)
print(predicted_sales)
```
