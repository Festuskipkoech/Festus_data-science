# K-Means Clustering

## Overview
K-Means is a popular unsupervised machine learning algorithm used for **clustering**. It partitions data into a predefined number of **K clusters** based on feature similarity.

---

## Key Concepts
1. **Clusters**:
   - Groups of similar data points.

2. **Centroids**:
   - Each cluster has a central point (centroid), which represents the average of all points in that cluster.

3. **Distance Metrics**:
   - Typically uses **Euclidean Distance** to measure the similarity between data points and centroids.

---

## Algorithm Steps
1. **Initialization**:
   - Randomly initialize `K` centroids.

2. **Assignment**:
   - Assign each data point to the nearest centroid.

3. **Update**:
   - Calculate the new centroids as the mean of the points in each cluster.

4. **Repeat**:
   - Iterate between assignment and update steps until centroids stabilize or a stopping criterion is met.

---

## Key Parameters
1. **K** (Number of Clusters):
   - Predefined by the user.
   - Can be determined using techniques like the **Elbow Method** or **Silhouette Analysis**.

2. **Max Iterations**:
   - Limits the number of times the algorithm will iterate.

3. **Initialization Method**:
   - Random Initialization or **K-Means++** (improves centroid initialization).

---

## Evaluation Metrics
1. **Inertia (Within-Cluster Sum of Squares)**:
   - Measures how tightly packed the points are within each cluster.
   - Lower inertia indicates better clustering.

2. **Silhouette Score**:
   - Measures how similar points are to their own cluster compared to others.
   - Range: [-1, 1] (Higher is better).

---

## Advantages
1. Simple to implement and computationally efficient.
2. Scales well to large datasets.
3. Works well with compact, spherical clusters.

---

## Limitations
1. Sensitive to:
   - Initial centroid placement (can result in local minima).
   - The value of `K` (requires manual tuning).
2. Struggles with non-spherical or overlapping clusters.
3. Outliers can significantly affect results.

---

## Applications
- Customer Segmentation
- Image Compression
- Document Clustering
- Market Segmentation

---

## Elbow Method
- Used to determine the optimal value of `K`.
- Plot **Inertia** vs. **K**.
- Look for the "elbow point" where the rate of decrease in inertia slows.

---

## Example Code in Python
```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Sample Data
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# K-Means Clustering
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Plotting Clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.legend()
plt.show()


# K-Means Clustering

## How K-Means Works

### 1. Overview
K-Means is an iterative algorithm that partitions data into `K` clusters by minimizing the **within-cluster sum of squares (inertia)**.  
Each data point is assigned to a cluster based on its distance to the cluster's centroid.

---

### 2. Steps of the Algorithm

#### Step 1: Initialize Centroids
- Select `K` initial centroids (randomly or using **K-Means++** for better initialization).
- In the code:
  - `init='k-means++'` improves initialization and ensures faster convergence.

#### Step 2: Assign Points to Clusters
- Assign each point to the nearest centroid based on **Euclidean Distance**.
- In the code:
  - `fit_predict(X)` assigns cluster labels to each data point.
  - The output `y_kmeans` contains these labels.

#### Step 3: Update Centroids
- Recalculate centroids as the mean position of all points in their respective clusters:
  \[
  \text{Centroid}_k = \frac{1}{N_k} \sum_{i \in k} X_i
  \]

#### Step 4: Repeat Until Convergence
- Repeat Steps 2 and 3 until:
  1. Centroids stabilize (i.e., positions do not change significantly).
  2. Maximum iterations (`max_iter`) are reached.

---

### 3. Key Concepts in the Code
- **K (Number of Clusters)**:
  - Predefined using `n_clusters`.
  - Specifies the number of groups to divide the data.
- **Distance Metric**:
  - K-Means uses **Euclidean Distance**:
    \[
    d = \sqrt{\sum_{j=1}^n (x_{ij} - c_{kj})^2}
    \]
    where \(x_{ij}\) is a data point, and \(c_{kj}\) is a centroid.
- **Centroids**:
  - Stored in `kmeans.cluster_centers_` after training.

---

### 4. Visualization
- **Data Points**:
  - Represented as scatter points, colored based on their assigned clusters (`c=y_kmeans`).
- **Centroids**:
  - Represented as large red markers (`kmeans.cluster_centers_`).
- **Clusters**:
  - Groups of points are formed based on proximity to centroids.

---

### 5. Stopping Criteria
K-Means stops iterating when:
1. **Centroids stop moving** (convergence).
2. **Maximum iterations** (`max_iter`) are reached.

---

### 6. Choosing the Optimal K
- The number of clusters (`K`) must often be determined manually.
- Techniques to find optimal `K`:
  - **Elbow Method**: Plot inertia vs. `K` and find the "elbow" point.
  - **Silhouette Score**: Measures the quality of clustering.

---

### 7. Advantages of K-Means
- Simple and easy to understand.
- Computationally efficient for small to medium-sized datasets.
- Effective for compact, spherical clusters.

---

### 8. Limitations of K-Means
- Sensitive to:
  - Initial centroid placement (may converge to local minima).
  - Outliers (can skew centroids).
- Assumes clusters are spherical and equally sized.
- Requires manual selection of `K`.

---

The distance between two data points can be calculated using various distance metrics, with the most common being **Euclidean Distance**.

### Euclidean Distance

In a 2D space, the Euclidean distance between two points `(x1, y1)` and `(x2, y2)` is calculated as:

\[
\text{Distance} = \sqrt{(x2 - x1)^2 + (y2 - y1)^2}
\]

In a higher-dimensional space (where each feature is a dimension), the formula becomes:

\[
\text{Distance} = \sqrt{(x2_1 - x1_1)^2 + (x2_2 - x1_2)^2 + \dots + (x2_n - x1_n)^2}
\]

Where:
- `x1_1, x1_2, ..., x1_n` are the features of the first data point.
- `x2_1, x2_2, ..., x2_n` are the features of the second data point.

### Example with Your Data Points

Letâ€™s use a simplified 2D example where the features are `height` and `weight`:

- **Data Point 1**: `(height = 5.4, weight = 130)`
- **Data Point 2**: `(height = 5.6, weight = 150)`

To calculate the Euclidean distance between these two points:

\[
\text{Distance} = \sqrt{(5.6 - 5.4)^2 + (150 - 130)^2}
\]

\[
\text{Distance} = \sqrt{(0.2)^2 + (20)^2}
\]

\[
\text{Distance} = \sqrt{0.04 + 400}
\]

\[
\text{Distance} = \sqrt{400.04} \approx 20.00
\]

This means that the **Euclidean distance** between these two data points is approximately `20.00`.

### In Higher Dimensions (e.g., with more features)

If you had more features (like `age`, `grades`, etc.), the Euclidean distance would take all those dimensions into account.

For instance, if we extend the data points to have 4 features:

- **Data Point 1**: `(age = 16, height = 5.4, weight = 130, grades = 90)`
- **Data Point 2**: `(age = 17, height = 5.6, weight = 140, grades = 85)`

The Euclidean distance would be:

\[
\text{Distance} = \sqrt{(17 - 16)^2 + (5.6 - 5.4)^2 + (140 - 130)^2 + (85 - 90)^2}
\]

\[
\text{Distance} = \sqrt{(1)^2 + (0.2)^2 + (10)^2 + (-5)^2}
\]

\[
\text{Distance} = \sqrt{1 + 0.04 + 100 + 25}
\]

\[
\text{Distance} = \sqrt{126.04} \approx 11.23
\]

This means the **Euclidean distance** between these two multi-dimensional data points is approximately `11.23`.

### Key Points:
- The **distance** between data points quantifies how "far apart" they are in the feature space.
- **K-Means clustering** uses this distance to assign data points to the nearest centroid.
- The Euclidean distance is the most common metric, but other distance measures (like Manhattan distance, cosine similarity, etc.) can also be used depending on the problem.

In summary, **distance** refers to the measure of similarity or dissimilarity between two data points in the feature space, and **Euclidean distance** is the most common way to calculate this distance in many clustering algorithms like K-Means.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
````

````python
# generate random data points
X,_=make_blobs(n_samples= 300, centers=4, random_state=42)
````

````python
# plot the data points
colors = ['red', 'blue', 'green', 'orange'] * 75
plt.scatter(X[:,0],X[:,1],s=50, c=colors)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title("Generated data points")
plt.show()
````

````python
# create K-means object with k=4 clusters
kmeans= KMeans(n_clusters=4, random_state=42)
# fit the k-means to the data
kmeans.fit(X)
````

````python
# Get the cluster labels assigned to each data point
labels = kmeans.labels_
# get cluster centres
centers= kmeans.cluster_centers_
# plot the datapoints with cluster assignment
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis')
# plot the cluster centers
plt.scatter(centers[:, 0], centers[:, 1], marker='o', c='red', s=200, alpha=0.7)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
````
