```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
police=pd.read_csv('/content/3. Police Data.csv')
police.head()
```

```python
police.info()
```

```python
police.describe()
```

```python
police.isnull().sum()
```

```python
# remove columns that only contain mising values
police.drop(columns='country_name', inplace=True)
```

```python
police.isnull().sum()
```

```python
# for speeding , were menor women stopped more often
police[police.violation == 'Speeding'].driver_gender.value_counts()
```

```python
# does gender affect who get searched more often
police.groupby('driver_gender').search_conducted.sum()
```

```python
police.search_conducted.value_counts()
```

```python
# mapping and data-type casting
# what is the mean stop duration?
police['stop_duration']=police['stop_duration'].map( {'0-15 Min' : 7.5, '16-30 Min' : 23,'30+ Min' :45} )
```

```python
police.head(2)
```

```python
# mean of stop_duration
police['stop_duration'].mean()
```
