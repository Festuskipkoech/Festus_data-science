# Random Forest

## Core Concepts
* Ensemble learning method combining multiple decision trees
* Each tree votes on prediction (majority for classification, average for regression)
* Uses bagging (bootstrap aggregating) to create diverse trees
* Random feature selection at each split improves diversity

## Key Differences from Single Decision Trees
* More robust and less prone to overfitting
* Higher accuracy through ensemble predictions
* Slower training and prediction time
* Less interpretable than single trees
* Better handling of high-dimensional data

## Working Mechanism
* Bootstrap Sampling:
   - Random sampling with replacement from training data
   - Each tree gets different training subset
   - Typically ~63.2% unique samples per tree (rest are duplicates)

* Random Feature Selection:
   - At each split, only consider subset of features
   - Default: sqrt(n_features) for classification
   - Default: n_features/3 for regression
   - Reduces correlation between trees

## Important Parameters
* n_estimators: Number of trees in forest
* max_depth: Maximum depth of each tree
* min_samples_split: Minimum samples for node split
* min_samples_leaf: Minimum samples at leaf node
* max_features: Features considered for best split
* bootstrap: Whether to use bootstrap sampling
* n_jobs: Number of parallel jobs for training

## Feature Importance
* Averaged across all trees in forest
* More reliable than single tree importance
* Calculated using:
   - Mean decrease in impurity
   - Mean decrease in accuracy
   - Permutation importance

## Advantages
* Reduces overfitting through averaging
* Handles high-dimensional data well
* Built-in feature importance ranking
* Handles missing values effectively
* Less parameter tuning than other algorithms

## Disadvantages
* Black box model (less interpretable)
* Larger memory requirement
* Slower prediction than single trees
* May still overfit with too many trees
* Not ideal for linear relationships

# Random Forest: Key Concepts and Basics

---

### 1. Overview

Random Forest is an **ensemble learning algorithm** that combines multiple decision trees to improve performance for both classification and regression tasks. It builds multiple trees during training and aggregates their outputs for predictions.

---

### 2. Key Features

- **Ensemble Learning**: Combines predictions from multiple decision trees.
- **Bagging** (Bootstrap Aggregating): Uses random subsets of data for training each tree.
- **Feature Subsampling**: Considers only a random subset of features at each split.
- **Aggregation**:
  - For classification: Uses majority voting from all trees.
  - For regression: Averages predictions from all trees.

---

### 3. Process of Random Forest

1. **Data Bootstrapping**:
   - Randomly sample the dataset with replacement to create multiple subsets for training trees.
2. **Training Decision Trees**:
   - Each tree is trained on a different bootstrapped dataset.
   - At each split, only a random subset of features is considered to find the best split.
3. **Prediction Aggregation**:
   - For classification: The class with the most votes across trees is the final prediction.
   - For regression: The mean value of all trees’ predictions is the output.

---

### 4. Advantages

- Reduces overfitting by combining multiple trees.
- Handles high-dimensional data well.
- Identifies the most important features.
- Robust to noise and missing data.

---

### 5. Disadvantages

- Computationally expensive due to training multiple trees.
- Less interpretable compared to a single decision tree.
- Higher memory usage to store all trees.

---

### 6. Hyperparameters

- **`n_estimators`**: Number of decision trees in the forest.
- **`max_depth`**: Maximum depth of each tree.
- **`max_features`**: Number of features to consider for splits (e.g., `'sqrt'`, `'log2'`).
- **`min_samples_split`**: Minimum number of samples needed to split a node.
- **`min_samples_leaf`**: Minimum number of samples in leaf nodes.
- **`bootstrap`**: Whether to use bootstrapping (default: True).

---

### 7. Important Concepts

#### 7.1. Bagging
- Combines predictions from multiple models trained on different subsets of the data.
- Reduces variance and increases robustness.

#### 7.2. Feature Subsampling
- Only a random subset of features is considered for splitting at each node.
- Reduces correlation between trees and improves generalization.

#### 7.3. Out-of-Bag (OOB) Error
- Estimates prediction error using data points not included in the bootstrap sample for each tree.
- Provides an unbiased estimate of accuracy.

#### 7.4. Feature Importance
- Measures how much each feature reduces impurity across all trees.
- Helps identify the most relevant predictors.

---

### 8. Applications

1. **Classification**:
   - Fraud detection.
   - Disease diagnosis.
   - Sentiment analysis.
2. **Regression**:
   - Predicting house prices.
   - Stock market analysis.
   - Weather forecasting.

---

### 9. Random Forest vs Decision Tree

| **Aspect**          | **Decision Tree**            | **Random Forest**                 |
|---------------------|------------------------------|------------------------------------|
| **Overfitting**     | Prone to overfitting         | Reduces overfitting (ensemble)    |
| **Accuracy**        | Moderate                    | High                              |
| **Speed**           | Fast                        | Slower (due to many trees)        |
| **Interpretability**| Easy                        | Harder                            |

---

### 10. Example in Python

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, max_depth=3, random_state=42)
rf.fit(X_train, y_train)

# Predict and evaluate
y_pred = rf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))


````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
# from sklearn.datasets import make_classificationfrom sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

````

````python
# load the dataset
data = load_breast_cancer()
````

````python
X= data.data
y=data.target
````

````python
# Split the data into training and testing
X_train, X_test, y_train,y_test=train_test_split(X,y, test_size=0.3,random_state=42)
````

````python
rf_classifier=RandomForestClassifier(n_estimators=100, random_state=42)
````

````python
# train the classifier
rf_classifier.fit(X_train,y_train)
````

````python
# predictions
predictions = rf_classifier.predict(X_test)
````

````python
import sklearn
accuracy = sklearn.metrics.accuracy_score(y_test, predictions)
print("Accuracy", accuracy)
````

````python
# get feature importances
importances= rf_classifier.feature_importances_
````

````python
# sort feature importances in descending order
indices = np.argsort(importances[:: -1])

# rearange features names do they match the sorted feature importances
feature_names= data.feature_names[indices]
````

````python
# create bar plot of feature importances
plt.figure(figsize=(10,6))
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices])
plt.xticks(range(X.shape[1]), feature_names, rotation=90)
plt.show()
````
