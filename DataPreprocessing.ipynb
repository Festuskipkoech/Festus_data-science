
## Step5: Handling Missing Values
We can apply 3 Strategies depending on the Data Type:

1. COUNTINOUS(NUMERICAL): The missing values can be replaced by either the MEAN, MEDIAN or KNN Imputer.

2. CATEGORICAL OR DISCRETE: The missing values can be replaced by the MODE(Most represented value in the Distribution) or KNN Imputer.

3. GENERAL: We can also can dropna() in our DataFrames and All the Missing Records will be Dropped from the DataFrame. e,g
 df.dropna()


```python

# Check All the Empty
df.isnull().sum()

# Fill the Life expectancy Empties
# Get the mean life expectancy
mean_exp = df['Life expectancy '].mean()

# Call the fillna() function
df['Life expectancy '].fillna(mean_exp, inplace = True)
df.isnull().sum()
```


## KNNImputer:

K-Nearest-Neighbor Imputer: Machine Learning Algorithms.
KNNImputer is a technique from the scikit-learn(machine learning) library used to handle missing values in datasets. It employs the k-nearest neighbors algorithm to impute (fill in) missing data points based on the values of their nearest neighbors. It is a good technique compared to the mean and the median because the value being filled is derived using k-nearest.
k

```python

from sklearn.impute import KNNImputer
imputer = KNNImputer()

for index in df.select_dtypes(include = 'number').columns:
    df[index] = imputer.fit_transform(df[[index]])

df.isnull().sum()
```

#HANDLING OUTLIERS


An outlier is a data point that significantly differs from the other points in a dataset. In statistics, outliers can be unusually high or low compared to the majority of data. They can result from variability in the data or indicate a measurement error or an unusual phenomenon.

IQR

It's used to understand the spread of the central portion of the data and to identify potential outliers.

```python
# Define to remove outlier fuction

# Define a Function to Remove Outliers
def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1
    minimum = Q1 - (1.5 * IQR)
    maximum = Q3 + (1.5 * IQR)

    return minimum, maximum
```

```python

# Detect Outlier for Life Expectancy
sns.boxplot(data=df, x='Life expectancy ')
plt.show()
```

```python
# Remove an outlier for life expectancy
minimum, maximum = remove_outliers(df['Life expectancy '])
print(minimum, maximum)
condition = (df['Life expectancy '] > minimum) & (df['Life expectancy '] < maximum)
df=df[condition]

sns.boxplot(data=df, x='Life expectancy ')
plt.show()
```
