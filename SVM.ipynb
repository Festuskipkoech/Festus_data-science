# Support Vector Machines (SVM)

## Basic Concepts
* Supervised learning algorithm for classification/regression
* Finds optimal hyperplane to separate classes
* Creates maximum margin between classes
* Support vectors are points closest to hyperplane
* Can handle linear and non-linear classification

## Key Components
### Hyperplane
* Decision boundary separating classes
* For 2D: line
* For 3D: plane
* For >3D: hyperplane
* Formula: w·x + b = 0 (w=weights, b=bias)

### Margin
* Distance between hyperplane and nearest points
* Maximizing margin helps generalization
* Soft margin allows some misclassification
* Hard margin requires perfect separation

### Support Vectors
* Points closest to decision boundary
* Determine position of hyperplane
* Only these points affect hyperplane
* Fewer support vectors = better generalization

## Kernels
### Linear Kernel
* K(x,y) = x·y
* Suitable for linearly separable data
* Fastest computation
* Good for high-dimensional data

### RBF (Gaussian) Kernel
* K(x,y) = exp(-γ||x-y||²)
* Most commonly used
* Good for non-linear data
* γ controls decision boundary flexibility

### Polynomial Kernel
* K(x,y) = (γx·y + r)^d
* Good for normalized data
* d = degree of polynomial
* More flexible than linear

### Sigmoid Kernel
* K(x,y) = tanh(γx·y + r)
* Similar to neural networks
* Less commonly used
* Can have numerical difficulties

## Important Parameters
### C (Regularization)
* Controls trade-off between margin and errors
* Larger C = less regularization
* Smaller C = more regularization
* Helps prevent overfitting

### Gamma (γ)
* Kernel coefficient for RBF/Polynomial/Sigmoid
* Controls influence of single training point
* Larger γ = closer points have more influence
* Smaller γ = points have more even influence

## Advantages
* Effective in high-dimensional spaces
* Memory efficient (uses support vectors)
* Versatile (different kernel functions)
* Good generalization with right parameters
* Works well with clear margin of separation

## Disadvantages
* Sensitive to feature scaling
* Long training time for large datasets
* Difficult to interpret
* Parameter tuning can be complex
* Not directly suitable for probabilistic prediction

# Support Vector Machines (SVM) Basics and Key Concepts

## What is SVM?
- **Support Vector Machine (SVM)** is a supervised machine learning algorithm used for classification and regression tasks.
- It aims to find the best decision boundary (or hyperplane) that separates different classes in the dataset.

---

## Key Concepts in SVM

### 1. Hyperplane
- A hyperplane is a decision boundary that separates data points of different classes.
- In **2D space**, it's a line. In **3D space**, it's a plane. In higher dimensions, it generalizes to an n-dimensional hyperplane.

---

### 2. Support Vectors
- **Support vectors** are the data points closest to the hyperplane.
- These points are critical in defining the position and orientation of the hyperplane.

---

### 3. Margin
- The **margin** is the distance between the hyperplane and the closest data points from each class.
- SVM tries to maximize this margin for better generalization.

---

### 4. Types of SVM
1. **Linear SVM**:
   - Used when the data is linearly separable.
   - Finds a straight hyperplane that separates the classes.

2. **Non-Linear SVM**:
   - Used when the data is not linearly separable.
   - Employs a kernel function to transform the data into a higher-dimensional space where it becomes linearly separable.

---

### 5. Kernel Functions
- Kernels map the original data into a higher-dimensional space to handle non-linear relationships.
- Common kernel types:
  - **Linear Kernel**: \( K(x, y) = x \cdot y \)
  - **Polynomial Kernel**: \( K(x, y) = (x \cdot y + c)^d \)
  - **Radial Basis Function (RBF) Kernel**: \( K(x, y) = \exp(-\gamma ||x - y||^2) \)
  - **Sigmoid Kernel**: \( K(x, y) = \tanh(\alpha x \cdot y + c) \)

---

### 6. Regularization (C)
- The **C parameter** controls the trade-off between maximizing the margin and minimizing classification errors.
  - A **small C** value creates a wider margin but may allow some misclassifications.
  - A **large C** value creates a narrower margin but aims to classify all training points correctly.

---

### 7. SVM for Classification
- SVM assigns a class label by determining on which side of the hyperplane a data point lies.

---

### 8. SVM for Regression (Support Vector Regression - SVR)
- SVM can also perform regression tasks.
- SVR attempts to fit the data points within a margin of tolerance (epsilon) around the hyperplane.

---

## Advantages of SVM
- Effective in high-dimensional spaces.
- Works well with a clear margin of separation.
- Robust to overfitting, especially in high-dimensional datasets.

---

## Disadvantages of SVM
- Computationally expensive for large datasets.
- Performance depends heavily on the selection of kernel and parameters.
- Not suitable for very noisy datasets.

---

## Key Hyperparameters in SVM
1. **Kernel**: Defines the type of hyperplane used to separate classes.
2. **C**: Regularization parameter.
3. **Gamma**: Controls the influence of a single training example in RBF kernels.
4. **Degree**: Degree of the polynomial kernel.

---

## Applications of SVM
- Text classification (e.g., spam detection).
- Image classification (e.g., facial recognition).
- Bioinformatics (e.g., cancer classification).
- Stock price prediction (regression tasks).

---

By understanding these basics and key concepts, you can apply SVMs effectively in various machine learning tasks.



### Example of Normalization
### Min-Max Normalization


**Definition**: Rescales values to the range [0, 1] using the formula:  

\[
x_{\text{normalized}} = \frac{x - \min(x)}{\max(x) - \min(x)}
\]

---

**Example**:  
Original data: [50, 60, 70]  
- **Min**: 50  
- **Max**: 70  

**Normalized Data**:  
\[
\begin{align*}
x_{\text{normalized}} &= \frac{x - \min(x)}{\max(x) - \min(x)} \\
\text{For 50: } &\frac{50 - 50}{70 - 50} = 0 \\
\text{For 60: } &\frac{60 - 50}{70 - 50} = 0.5 \\
\text{For 70: } &\frac{70 - 50}{70 - 50} = 1
\end{align*}
\]

Result: [0, 0.5, 1]


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
````

````python
# generating a synthentic dataset
np.random.seed(0)
X=np.r_[np.random.randn(20,2) - [2,2], np.random.randn(20,2) + [2,2]]
y=[0] * 20 + [1] *20
````

````python
# visualize the dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
````

````python
# training the model
clf=svm.SVC(kernel='linear')
clf.fit(X,y)
````

````python
# plotting the decision boundary
ax=plt.gca()
xlim=ax.get_xlim()
ylim=ax.get_ylim()

# create a grid to evaluate model
xx=np.linspace(xlim[0], xlim[1], 30)
yy=np.linspace(ylim[0], ylim[1], 30)
YY, XX= np.meshgrid(yy,xx)
xy=np.vstack([XX.ravel(), YY.ravel()]).T
Z= clf.decision_function(xy).reshape(XX.shape)

# plot decision bondary
ax.contour(XX, YY, Z, colors='k', levels=[-1,0,1], alpha=0.5, linestyles=['--', '-', '--'])

# plotting support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
````

````python
# make prediction
np.random.seed(1)
X_test =np.random.randn(10,2)
y_pred = clf.predict(X_test)

print("Predictions", y_pred)
````
