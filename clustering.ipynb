```python
data.dropna(inplace=True)
data.isnull().sum()
# drops any missing cells and line checks if the dropping has occurred
```

```python
array = data.values
array
#convertion from a dictionary to a list an array allows input of different data types in this case int and str
```

```python
#:, selects the entire list then we use indexes to elliminate the categorical data type
# 2:5 due to minus 1 rule
# we assign it a variable x because array is a reserved word
x = array[:, 2:5]
x
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10, random_state=42)
# random_state=42 ensures consistency when the code is run a number of times
model.fit(x)
```

```python
# means = model.cluster_centers_
clusters=pandas.DataFrame(model.cluster_centers_, columns=['age','income', 'score'])
clusters

# gets the mean of the clusters
```

```python
data.head()
```

```python
data['cluster']= model.labels_
data.head()
```
