# Principal Component Analysis (PCA): Core Components

Principal Component Analysis (PCA) is a dimensionality reduction technique widely used in data analysis and machine learning. Below are the core components and key concepts of PCA.

---

## 1. **Purpose of PCA**
- PCA reduces the number of dimensions in a dataset while retaining as much variance (information) as possible.
- It simplifies high-dimensional data into a smaller set of components, which are linear combinations of the original features.
- Common use cases include:
  - **Data visualization** (reducing data to 2D or 3D for plotting).
  - **Noise reduction**.
  - **Feature extraction** and **decorrelation**.

---

## 2. **Key Concepts of PCA**

### a) **Variance and Covariance**
- **Variance** measures the spread of data points within a feature.
- **Covariance** measures the relationship between two features.
  - A positive covariance means the two features increase together.
  - A negative covariance means one feature increases while the other decreases.

### b) **Eigenvalues and Eigenvectors**
- **Eigenvalues** represent the amount of variance captured by each principal component.
- **Eigenvectors** indicate the direction of the principal components in the feature space.
- In PCA, we compute the eigenvalues and eigenvectors of the **covariance matrix** of the dataset.

### c) **Principal Components**
- Principal components are new axes (dimensions) created by linear combinations of the original features.
- The first principal component captures the maximum variance in the dataset.
- Each subsequent component is orthogonal (perpendicular) to the previous ones and captures the remaining variance.

---

## 3. **Steps in PCA**

1. **Standardize the Data**:
   - Scale the features so they have a mean of 0 and standard deviation of 1 (important to ensure fair contribution of all features).

2. **Compute the Covariance Matrix**:
   - The covariance matrix summarizes relationships between the features.

3. **Calculate Eigenvalues and Eigenvectors**:
   - Eigenvectors define the directions of the principal components.
   - Eigenvalues determine the magnitude of variance explained by each component.

4. **Select Principal Components**:
   - Choose the top k principal components based on the largest eigenvalues.

5. **Transform the Data**:
   - Project the original data onto the new principal components to create a reduced dataset.

---

## 4. **Explained Variance**
- **Explained Variance** measures how much information (variance) is retained by each principal component.
- The **explained variance ratio** is the proportion of the total variance explained by a particular principal component.
- A **scree plot** is often used to visualize the cumulative explained variance and decide the number of components to retain.

---

## 5. **Advantages of PCA**
- **Dimensionality Reduction**: Reduces the number of features, improving computational efficiency.
- **Feature Decorrelation**: Removes redundancy by making features orthogonal (uncorrelated).
- **Visualization**: Simplifies high-dimensional data for better understanding and insights.

---

## 6. **Limitations of PCA**
- **Loss of Interpretability**: Principal components are linear combinations of original features, making them harder to interpret.
- **Linear Assumption**: PCA assumes linear relationships between features and may not work well with non-linear data.
- **Sensitivity to Scaling**: PCA is sensitive to the scale of the data, making standardization a crucial preprocessing step.

---

## 7. **Applications of PCA**
- **Image Compression**: Reduces image dimensions while preserving key visual information.
- **Gene Expression Analysis**: Identifies patterns in high-dimensional biological data.
- **Finance**: Reduces market data dimensions to analyze trends.

---

## Conclusion
PCA is a powerful tool for dimensionality reduction and exploratory data analysis. By transforming data into orthogonal components, it simplifies complex datasets and retains the most important information for downstream analysis.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
```

```python
iris= pd.read_csv('/content/iris.data.csv', header=None)
```

```python
iris.head()
```

```python
X = iris.iloc[:, :-1].values
y=iris.iloc[:, -1].values
```

```python
pca =PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
X_pca
```

```python
le=LabelEncoder()
v=le.fit_transform(y)
plt.scatter(X_pca[:, 0], X_pca[:,1], c=v, cmap='viridis')
plt.xlabel('Priicipal component 1')
plt.ylabel('Principal component 2')
plt.show()
```
