```python
import pandas as pd
```

```python
car=pd.read_csv('/content/Project+2+-+Cars+Dataset.csv')
car.head(2)
```

```python
car.info()
```

```python
car.shape
```

```python
car.isnull().sum()
```

```python
# we fill the null values with the mean of that column
column_fill=['Cylinders','Make','Model','Type','Origin','DriveTrain','MSRP','Invoice','EngineSize','Horsepower','MPG_City','MPG_Highway','Weight','Wheelbase','Length']
for column in column_fill:
  if car[column].dtype in['float64', 'int64']:
      car[column]=car[column].fillna(car[column].mean(), inplace=True)
# for one column only
# car['Cylinders'].fillna(car['Cylinders'].mean(), inplace=True)
```

```python
car.isnull().sum()
```

```python
car.dropna(inplace=True)
```
